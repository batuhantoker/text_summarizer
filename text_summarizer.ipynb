```python

import fitz  # this is pymupdf

from transformers import PegasusForConditionalGeneration, PegasusTokenizer

from transformers import BartTokenizer, BartForConditionalGeneration
tokenizer_model = PegasusTokenizer.from_pretrained("google/pegasus-xsum")
loaded_model = PegasusForConditionalGeneration.from_pretrained("google/pegasus-xsum")

model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
from transformers import pipeline 
summarizer = pipeline("summarization")
```

```python
# Generate text from article PDF
with fitz.open("/content/paper1.pdf") as doc:
    text = ""
    for page in doc:
        text += page.get_text()
```

```python
#text as input
text = "t'undamental to understanding the way a human controls motion is the recognition that the ~uman is an adaptive, self - optimizin~ system . ~ iven the opportunity the human operator will ;)ptirnize his per formance on a tasL ;~ehavior may be adjusted to minimize effort required , time taken , or any combination of these and o th e r costs . The optimiZAtion criterion depends upon the task and is, in fac t , an implicit statement of the task ." #@param {type:"string"}
```

```python
# Summary method 1

tokens = tokenizer_model(text, truncation=True, padding="longest", return_tensors="pt")
summary = loaded_model.generate(**tokens)
tokenizer_model.decode(summary[0])
```

```python
# Summary method 2
inputs = tokenizer([text],truncation=True, max_length=1024, return_tensors="pt")

# Generate Summary
summary_ids = model.generate(inputs["input_ids"], num_beams=20, max_length=500)
tokenizer.batch_decode(summary_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]

```

```python
# Summary method 3
#run the model
summarized = summarizer(text, min_length=25, max_length=200)

# Print summarized text
print(summarized)
```
